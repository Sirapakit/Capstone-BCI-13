```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_array = np.load('./main_join_file_array_T8-P8.npy')
```

```python
print(data_array.shape)
print(np.max(data_array))
print(np.min(data_array))
```

```python
seizure_info_start = np.array([10196, 12267, 52132, 55015, 62920, 71390, 90925]) / (3600)
seizure_info_stop =  np.array([10236, 12294, 52172, 55066, 63010, 71483, 91026]) / (3600)
vline = np.append(seizure_info_start, seizure_info_stop)
vline.sort()
print(vline * 3600)
```

```python
# Rough Plot
# Plot join_file
t = np.linspace(0, data_array.shape[0]/(3600*256), data_array.shape[0], False)

plt.subplot(211)
plt.title('Raw concat file')
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.plot(t, data_array)

# Normalise 
data_array = (data_array-np.min(data_array))/(np.max(data_array)-np.min(data_array))

plt.subplot(212)
plt.title('Normalise file')
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.plot(t, data_array)

plt.tight_layout()
plt.show()
```

```python
# First 
plt.subplot(411)
plt.title('10 mins before')
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.xlim([seizure_info_start[0] - (600/3600) , (seizure_info_stop[0])])
plt.plot(t, data_array)

# Second
plt.subplot(412)
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.xlim([seizure_info_start[1] - (600/3600) , (seizure_info_stop[1])])
plt.plot(t, data_array)

# Third
plt.subplot(413)
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.xlim([seizure_info_start[2] - (600/3600) , (seizure_info_stop[2])])
plt.plot(t, data_array)

# Fourth
plt.subplot(414)
plt.vlines(vline, np.min(data_array), np.max(data_array), colors='red', ls='--', lw=1)
plt.xlim([seizure_info_start[3] - (600/3600) , (seizure_info_stop[3])])
plt.plot(t, data_array)

plt.tight_layout()
plt.show()
```
