```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path1= r'../convert-all-channel-edf/8bands-chb12-Part1.npy'
data1 = np.load(data_path1)
print(data1.shape)

data_path2 = r'../convert-all-channel-edf/8bands-chb12-Part2.npy'
data2 = np.load(data_path2)
print(data2.shape)

data_chb12 = np.hstack((data1, data2))
print(data_chb12.shape)

del data_path1, data_path2, data1, data2
```

```python
# data_selected_band = data_chb12[88:95+1,:]
data_selected_band = data_chb12[88,:]


data_class_0 = data_selected_band[4213: 6013]
plt.subplot(311)
plt.plot(data_class_0)
plt.ylim(0,0.5)

data_class_1 = data_selected_band[6013: 6913]
plt.subplot(312)
plt.plot(data_class_1)
plt.ylim(0,0.5)

data_class_2 = data_selected_band[6977: 6997]
plt.subplot(313)
plt.plot(data_class_2)
plt.ylim(0,0.5)
plt.show()
```

```python
sampling_rate = 256
NFFT = 256

fig, ax = plt.subplots()
Pxx, freqs, bins, im = plt.specgram(data_class_0, NFFT=NFFT, Fs = sampling_rate, cmap='seismic', noverlap=NFFT/2)
fig.colorbar(im, fraction=0.046, pad=0.04).set_label('Intensity [dB]')
# plt.ylim([0, 25])
plt.title("Spectrogram")
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
```
